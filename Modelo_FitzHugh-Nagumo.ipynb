###FitzHughâ€“Nagumo model (FHN)

```python
a = 0.8
b = 0.7
tau = 12.5
I_ext = 0.5

T = 200
h = 0.1
M = T/h

t = np.arange(0.0, T+h, h)

v = [] # elevation of membrane voltage
w = [] # recovery variable

v.append(0)
w.append(0)

for i in range(int(M)):
  _v = v[i] - (pow(v[i],3)/3) - w[i] + I_ext
  v.append(_v*h + v[i])

  _w = (v[i+1] + a - b*w[i]) / tau
  w.append(_w*h + w[i])


# V x t 
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1)

ax.spines['bottom'].set_position('center')

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.set_xlabel('Time (s)')
ax.xaxis.set_label_coords(1.05, .52)

plt.plot(t, v)

plt.ylabel("Voltage (V)")
plt.show()


# w x v
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1)

ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

ax.set_xlabel('v in volts')
ax.xaxis.set_label_coords(1.08, .52)
ax.set_ylabel('w in miliamps', rotation='horizontal')
ax.yaxis.set_label_coords(.5, 1.02)


plt.plot(v, w)

plt.show()
```
